```python
# 범주형 데이터 결측치 처리 함수 
import random
def category_columns_replace(df):
    # object인 column들(카테고리)만 뽑기
    category_columns_object = df.select_dtypes(include='object').columns

    # 결측치 0개 초과 10000개 미만의 데이터를 대상으로 랜덤하게 결측치 대체
    for column in category_columns_object:
        if df[column].isna().sum() > 10000 or df[column].isna().sum() == 0:
            continue
        unique_columns = df[column].loc[df[column].isna()==False].unique()
        df[column] = df[column].apply(lambda x : random.choice(unique_columns) if pd.isna(x) else x)
        
    return df
```
